## Question-1: Guess the Scrambled Word

## Algorithm:
1. Start a try block which accepts a file input name from the user.
2. Accept game input in the form of 'Y' or 'N' from the user, where cases do not matter.
3. If the game input is 'Y',set a flag game_status as True.
4. Read all the lines of the file and split using ',' to get the meanings and words.
5. Store the words and their meanings after splitting in a dictionary.
6. Choose a word randomly from the dictionary.
7. Scramble the characters of that word
8. Print the scrambled word on the screen asking user to unscramble.
9. Initiate counters to count the number of trials user has used and number of times they asked for meaning
10. Number of attempts the user has should be equal to the number of characters the word has
11. Start a while loop till the try count is equal to the lenght of the word.
12. Accept an answer input from the user
13. If the user enters a question mark ('?'), print the meaning of the word and do not count it as an attempt
14. If the user enters '?', give a warning saying the user already requested for meaning and next trial would be counted as an attempt.
15. If the user enters '?', more than twice, count every trial as an attempt.
16. If the user did not enter the correct input, count it as an attempt and give another chance.
17. If the user gets it wrong again and it is the last attempt, print the same on the screen.
18. If the user gets it right, print "Congrats you got it right"
19. If all the attempts are done, print 'Game Over'
20. Once all attempts are done, ask the user again if they want to play
21. Keep playing until the user says 'N'
22. If the user says 'N', exit the game and print 'Goodbye!'
23. Stop the program.

## Instructions for running the program:
- Upon running the program, enter the file name as it is with an extension .txt. File input name is case sensitive.
- When prompted if you want to play, type ONLY 'Y', 'N','y' or 'n'. BUT NOT 'YES' or 'NO'. Else the program will terminate.
- The input file needs to have words and meanings separated only by a comma ',' as stated in the problem statement.

## Expected results:
- If the user gets the word right, then 'Congrats, you got it!' is printed.
- If the user does not get the word right, then "You got it wrong, try again" is printed.
- If the user does not get the word right and it is the last attempt, then "You got it wrong, last attempt remaining." is printed
- If the user asks for meaning by entering '?', then the meaning is printed.
- If the user asks for meaning TWICE by entering '?' two times, "Warning: You have asked for meaning before. Next time would count as an attempt." a warning is printed.
- If the user asks for meaning more than TWICE, "You have asked for meaning more than twice, counting this as an attempt" is printed
- If all the attempts are exhausted, then 'Game Over' is printed
- If the user says 'n' when asked for if they want to play, then the program is terminated and 'Goodbye!' is printed.
- Along with the above statements, the user's previous inputs are also printed to let user verify their spellings incase.

```python
# Importing random function
import random
try:
    # Accepting file input name from the user, containing the words and their meanings separated by ':'
    file_name = input("Enter the name of the file: ")
    # Asking user if they want to play the game or not
    # Input needs to be only 'Y' or 'N'. Cases dont matter
    # DO NOT ENTER 'YES' OR 'NO'
    game_input = input("Do you want to play this game? ").lower().strip()
    if game_input == 'y':
        game_status = True
        while game_status == True:
            word_dict = {}
            # Opening the words file
            text_file = open(file_name)
            lines = text_file.readlines()
            for line in lines:
                # Looping through the meanings, where the word and meanings are separated by ','
                words = line.split(',')
                # Splitting at ',' to separate the meanings
                # Creating a dictionary with words and meaning
                word_dict[words[0]] = words[1].strip()
            # Randomly choosing a word from the above dictionary and converting to lower case for easy analysis
            random_w = random.choice(list(word_dict))
            random_word = random_w.lower()
            # Scrambling the word picked up randomly
            jumbled = random.sample(random_word, len(random_word))
            jumbled_word = ''.join(jumbled).lower()
            word_len = len(jumbled_word)
            # Printing the scrambled word for the user to guess
            print("The scrambled word is: ",jumbled_word)
            # Setting the counter values to '0'
            try_count = 0
            q_mark = 0
            # Creating an empty list to capture all the user inputs
            user_input = []
            # Number of attempts a user gets is the length of the word
            trials_left=word_len
            # While the user still has attempts left, loop continues.
            while try_count != word_len:
                print("********Number of trials left********: ",trials_left)
                # Accepting input from the user and converting to lower case
                answer_input = input("Enter the answer [or ? for the meaning]: ").lower().strip()
                # Appending the response to the list, for user to verify
                user_input.append(answer_input)
                print("Your previous inputs are: ",user_input)
                # If the user asks for meaning
                if answer_input == "?":
                    # Counter is increased to 1
                    q_mark += 1
                    # If the user asked for meaning more than twice, then the number of attempts left is reduced by 1
                    if q_mark>=2 and user_input.count("?") > 2:
                        print("You have asked for meaning more than twice, counting this as an attempt")
                        try_count+=1
                        trials_left-=1
                        continue
                    # If its the second time a user has asked for meaning, then a warning is given
                    # Number of attempts are not reduced here
                    elif user_input.count("?") > 1:
                        q_mark += 1
                        print("Warning: You have asked for meaning before. Next time would count as an attempt.")
                        continue
                    # Else the meaning of the word is printed
                    else:
                        print("The meaning is: ",word_dict[random_w])
                        continue
                # If the user guessed it right, they won the game and loop is broken
                elif answer_input == random_word:
                    print("Congrats, you got it!")
                    break
                # If the user input does not match and when its the last attempt
                elif (answer_input != random_word) and (try_count == word_len-2):
                    try_count+=1
                    trials_left-=1
                    print("You got it wrong, last attempt remaining.")
                    continue
                # This block is executed when the last attempt is exhausted
                elif (answer_input != random_word) and (try_count == word_len):
                    try_count+=1
                    trials_left-=1
                    continue
                # When in any case the word does not match, then this block is executed
                elif answer_input != random_word:
                    try_count+=1
                    trials_left-=1
                    print("You got it wrong, try again")
                    continue
            # When all the attempts are exhausted then this statement is printed
            print("Game over!")
            # Asking user if they want to play the game again
            game_input2 = input("Do you want to play this game again? ").lower().strip()
            # If it is a no, then the game is terminated
            if game_input2 == 'n':
                print("Goodbye!")
                game_status == False
                break
    # After the game is terminated
    elif game_input == 'n':
        print("Goodbye!")
    else:
        # If the user does not enter a 'Y' or 'N', game is terminated
        print("You did not enter 'N' or 'Y' hence exiting game!")
except Exception as e:
    print(e)
    # If the file name does not exist in the directory then this exception is raised
    print("File does not exist in the directory")
```

## Question-2: Cosine Similarity between poems

## Algorithm:
1. Start by accepting a file name from the user.
2. Accept an input poem from the user.
3. Open the file and read all the lines
4. Split the lines in the file at ':' to separate the poem from the name of the poet
5. Split at '/' and append all the lines of the poem into a list.
6. Append all the lists of all the poems into one list.
7. Split all the words in a poem and append them into a new word list.
8. Create a dictionary with the words and their count and append this dictionary in a new list called dict_list
9. Read the poem input by the user, split at '/' and append it into a list.
10. Split all the words in the user input and append into a new list called user_dict 'n' times, where 'n' is the number of lines in the poet file 
11. Find the difference of words between user input and the file input and append all these words in a list called main_user_list
12. Find the difference of words between file input and the user input and append all these words in a list called main_poet_list
13. If a word exists in poet file but not in user input, update the user dictionary with key being the word and value being '0'
14. If a word exists in user file but not in poet input, update the poet dictionary with key being the word and value being '0'
15. Calculate the number of '/' in each poem and append with the respective count in the each of the poem dictionary
16. Sort the dictionary based on the ascending order of the keys.
17. Pull all the values from each dictionary and store in a list.
18. Compare the poet vector with the user vector and calculate the cosine distance for each poet.
19. Print the cosine distance values of all the poems to the user poem.
20. Print the poem with the maximum cosine distance as the most similar poem.
21. If the maximum cosine distance is equal to '0', then print that 'Poems are not related at all'.
22. Catch the exception in case of a zero division error and print that 'Poems are not related at all'.
23. Raise an exception if the file does not exist in the directory.
24. Stop the program

## Instructions for running the program:
- Upon running the program, enter the file name as it is with an extension .txt. File input name is case sensitive.
- Only one poem is expected as an input and the poem lines need to be separated by a forward slash '/'

## Expected results:
- Cosine distance is printed for all the poets
- The poem with the maximum cosine distance is considered as the closest or most similar to the user poem
- If the poems are not related then, 'Poems are not related at all' is printed.
- Example:
- - Enter the name of the file: aa.txt
- - Input your poem delineated by “/” for each line: The world is too much with us
- - Cosine distances are below:
- - William Wordsworth :  0.3535533905932738
- - William Shakespeare :  0.23362835924302425
- - Robert Frost :  0.21309984068573073
- - CLOSEST POEM IS:  William Wordsworth: The world is too much with us; late and soon,/Getting and spending, we lay waste our powers;/Little we see in Nature that is ours;/We have given our hearts away, a sordid boon!


## Assumption:
- ASSUMPTION: Maximum Cosine Distance, Maximum Similarity
- Reference to the assumption: https://www.geeksforgeeks.org/cosine-similarity/

```python
try:
    # Accepting name of the file from user for text analysis
    file_name = input("Enter the name of the file: ")
    # Accepting input poem differentiating the lines with a '/'
    user_input = input("Input your poem delineated by “/” for each line: ")
    text_file = open(file_name)
    # Reading the lines of the file and adding them into a list
    lines = text_file.readlines()
    # 'n' denotes number of lines in the file, which is the number of different poets in the file
    n = len(lines)
    # Creating empty dictionaries and list variables for later part of the code
    word_list = []
    word_dict = {}
    all_lines = []
    main_lines = []
    poet_lines = []
    dict_list = []
    poet_len = []
    ########################### Part-1: FINDING NUMBER OF SLASHES IN A POEM ################
    # This is the part-1 of the code, where we are calculating the number of slashes in the user input
    a_user_input = user_input
    b_user_input = a_user_input.split("/")
    # Subtracting 1 from len(b_user_input) because we need to find the number of slashes and not number of lines
    # Number of slashes in a poem are one less than the number of lines
    n_lines_user = len(b_user_input) - 1    
    ########################### Part-2: WORKING WITH FILE INPUTS #########################
    # This is part-2 of the code, where we are analysing the poems in the file which user provided
    # Creating empty lists to store poet names
    poet_name = []
    poet_names = []
    for line in lines:
        # Looping through the list 'lines 'created above after opening the file 
        main_lines = []
        # all_lines is a list which contains only the poem and removes poet name
        # Converting to lower cases for easy analysis, splitting at ":" for removing poet name
        # Further splitting at '/' to append all the lines of the poem in a single list
        poet_name = line.strip().split(':')[0]
        poet_names.append(poet_name)
        all_lines = line.lower().strip().split(':')[1].split('/')
        poet_len.append(len(all_lines))
        for i in all_lines:
            main_lines.append(i.strip())
        # Poet lines is a list containing lists having one poem each as their elements
        poet_lines.append(main_lines)
    for i in range(n):
        # Looping through above lists for splitting the words from the poem lines
        b = []
        for j in poet_lines[i]:
            a = []
            a = j.split()
            b.extend(a)
        # word_list is a list containing lists for each poem which have all the words separated by comma 
        word_list.append(b)
    # To start the index from zero, we are setting the counter value of poet_no as -1
    poet_no = -1
    # looping through each word in the words list to count the number of occurences
    for i in word_list:
        # Incrementing the poet index by 1 when we loop through the poets in sequence
        poet_no += 1
        word_dict = {}
        for j in i:
            # Counting the occurences of words in a list and adding it to the dictionary
            word_dict[j] = i.count(j)
        # Once dictionary for each poet is ready, we append it to the dict_list[] in the index same as the poet
        ### DICT LIST IS THE MAIN DICTIONARY WITH COUNT FROM THE FILE INPUT
        dict_list.insert(poet_no,word_dict)
    ##################################### Part-3: WORKING WITH USER INPUT #############################
    # This is part-3 of the code, where we are analysing the poems which user provided
    # Firstly, converting to lower cases for easy analysis, then stripping to remove any end line or other characters
    # Further splitting at '/' to append all the lines of the user poem in a single list
    user_lines = user_input.lower().strip().split('/')
    user_words = []
    user_dict = {}
    for i in user_lines:
        # Looping through each line and appending the words separately into a list user_words
        c = []
        c = i.split()
        user_words.extend(c)
    for i in user_words:
        # Counting the number of occurences of a word in the list and appending it into a dictionary
        user_dict[i] = user_words.count(i)
    ##################################### Part-4 : WORKING WITH BOTH DICTS #############################
    # Creating empty lists required later
    user_dict_new = []
    user_diff_list = []
    poet_diff_list = []
    main_user_list = []
    main_poet_list = []
    for i in range(n):
        # Looping through one line at a time
        # Creating a duplicate list having the word count dictionaries
        user_dict_new.insert(i,user_dict)
        # Finding a difference of words between user words and the file words
        user_diff_list.insert(i,(set(user_dict_new[i])-set(dict_list[i])))
        # Finding a difference of words between file words and the user words
        poet_diff_list.insert(i,(set(dict_list[i])-set(user_dict_new[i])))
        # Inserting the difference of words into new lists, one for file and one for user input
        main_user_list.insert(i,list(user_diff_list[i]))
        main_poet_list.insert(i,list(poet_diff_list[i]))
        # Since some user words are not there in the file, we are setting the value of it to '0'
        for j in main_user_list[i]:
            dict_list[i][j] = 0
    # Finding the length of the list for looping
    ln = len(main_poet_list)
    # Creating a duplicate dictionary to user_dict for adding the final word count
    user_dictt = []
    for i in range(ln):
        # adding empty dictionaries equal to the number of poets
        user_dictt.insert(i,{})
    for c in range(ln):
        # Since some file words are not there in the user poem, we are setting the value of it to '0'
        for x in main_poet_list[c]:
            user_dictt[c][x] = 0
        ### APPENDING THE NUMBER OF SLASHES IN THE FILE AND USER DICTIONARIES
        ### Number of slashes (represents number of lines) is also an important parameter for finding the similarity
        user_dictt[c]["/"] = n_lines_user # Calculated earlier
        dict_list[c]["/"] = poet_len[c] - 1 # poet_len represents number of lines, subtracting 1 gives number of slashes 
    for i in range(n):
        # Copying the updated elements to new list
        user_dictt[i].update(user_dict_new[i])
        
#         # Below code can be used to verify if the number of elements in the poem dictionary and user dictionary are same
#         # Both needs to be equal for cosine vector calculation
#         print("LEN DICT LIST IS: ",len(dict_list[i]))
#         print("LEN USER DICT IS: ",len(user_dictt[i]))
    
#     # Below two lines can be used to print the words and their respective counts
#     print("@@@@@@@POET DICT IS: ",dict_list)   
#     print("******USER DICT IS: ",user_dictt)
    
    #################################### Part-5: WORKING WITH VECTORS ################################
    # DICT_LIST AND USER_DICTT ARE MAIN LISTS WITH FINAL WORDS
    # creating empty lists to append the vectors with the word counts
    poet_vector = []
    user_vector = []
    # Looping through all the dictionaries in all the lists
    for i in range(n):
        # sorted() function arranges the keys in dictionaries as per their ascending order, and add them into a list
        # Below lines can be used for printing the sorted dictionaries with their count, for verifying the vector 
#         print("&&&&&&&&&&&&&&&&&&&&&: ",sorted(dict_list[i].keys()))
#         print("%%%%%%%%%%%%%%%%%%%%%: ",sorted(user_dictt[i].keys()))
        poet_v = []
        user_v = []
        for key in sorted(dict_list[i].keys()) :
            # Adding the word count of user file as elements of a list into the above lists, in ascending order
            poet_v.append(dict_list[i][key])
        for key in sorted(user_dictt[i].keys()) :
            # Adding the word count of user input as elements of a list into the above lists, in ascending order
            user_v.append(user_dictt[i][key])
        # Appending the above lists into a main vector list, arranged in their ascending order
        # The count of '/' is the first element in the vector
        poet_vector.insert(i,poet_v)
        user_vector.insert(i,user_v)
#     # Below two lines can be used for printing the vector count, for verrification if required
#     print("ppppppppppppppppppppPOET VECTOR IS: ",poet_vector)
#     print("uuuuuuuuuuuuuuuuuuuuUSER VECTOR IS: ",user_vector
    ###################################### COSINE FORMULA #####################################
    ### Cosine formula used is same as mentioned in the assignment file
    ### USER_VECTOR AND POET_VECTOR are the final lists containing the list of vectors from each poem
    # Setting the numberator, denominator and other variables required for the cosine calculation to '0'
    NR = 0
    DR = 0
    sum_Ai_square = 0
    sum_Bi_square = 0
    # Creating an empty list to append all the cosine distances at last
    cosine_dist = []
    for i in range(n):
        # Looping through each list containing vectors
        for j in range(len(user_vector[i])):
            # Looping through each vector
            # Calculating the numerator and denominator as per the formula
            # NUMERATOR:
            NR += user_vector[i][j]*poet_vector[i][j]
            sum_Ai_square += (user_vector[i][j]**2)
            sum_Bi_square += (poet_vector[i][j]**2)
        # Calculating the denominator as per the formula
        DR = ((sum_Ai_square)*(sum_Bi_square))**(0.5)
        # Appending all the cosine distances into a list
        # Zero Division error handled in the exception segment.
        cosine_dist.append(NR/DR)
    
    # Below print line can be used to print all the cosine distances.
    print("Cosine distances are below:")
    for i in range(n):
        
        print(poet_names[i], ': ',cosine_dist[i])
    
    # Finding the largest cosine distance
    ### ASSUMPTION: Maximum Cosine Distance, Maximum Similarity
    ### Reference to the assumption: https://www.geeksforgeeks.org/cosine-similarity/, 
    # The corresponding poem would be the closest to the user input
    max_cosine_dist = max(cosine_dist)
    if max_cosine_dist == 0.0:
        # Maximum cosine distance would be equal to '0.0' if the lines are not related at all.
        print("The poems are not related at all!")
    else:
        # Finding the poem index of the maximum distance
        poem_index = cosine_dist.index(max_cosine_dist)
        # Printing the closest poem
        print("CLOSEST POEM IS: ",lines[poem_index])

except ZeroDivisionError:
    # Zero Division error is observed when the denominator is zero
    # This arises in only once scenario, when all the vector elements of either file or user are zero
    # When all the vectors are zero, it means either there is no input provided or the provided input has only spaces
    # Therefore, since it does not contain any words, the input poem is not related to the poems in the file at all.
    print("The poems are not related at all!")
    
except Exception as e:
    print(e)
    # If the file name which user entered does not exist in the directory, this exception is raised.
    print("File not found on the directory")
```
